```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('vgsales-12-4-2019.csv')
df
```

```python
df
```

```python
df['Platform'] = df['Platform'].apply(lambda x: x.upper())
df
```

```python
df['Platform'].loc[(df['Platform'] == "NS")] = "Switch"
df['Platform'].loc[(df['Platform'] == "PSV")] = "VITA"
df['Platform'].loc[(df['Platform'] == "XB")] = "XBOX"

```

```python
df
```

```python
score_df = pd.read_csv('result.csv')
score_df
```

```python
games_df = pd.merge(df, score_df,  how='left', left_on=['Name','Platform'], right_on = ['name','console'])
```

```python
games_df
```

```python
games_df.columns
```

```python
games_df = games_df[games_df['metascore'].notnull()]
games_df = games_df[games_df['ESRB_Rating'].notnull()]
games_df = games_df[games_df['userscore'] != 'tbd']
games_df = games_df[games_df['Global_Sales'].notnull()]
games_df = games_df[games_df['Global_Sales'] > 0]

games_df
```

```python
games_df['console'].value_counts()
```

```python
games_df
```

```python
games_df.to_csv('games_clean_df.csv')
```
